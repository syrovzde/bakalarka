## Key findings
- it is not easy to beat a bookmaker, even without a margin!
    - you still need some (weaker) form of advantage!
- i.e. the overall EV of the game can be negative!
    - thus it does not hold that oposite directions will give you opossite returns etc. (they can just simply all be negative)
- it is generally advantageous for the bookmaker to be slightly biased towards the outsider (the value is on the favorite)
- this particular bookmaker is (almost) exactly as such
    - it is thus profitable to simply bet on the favorite/home
- we should now remove the bias to make the problem interesting for students (otherwise it's trivial and not valuable)

# Artificial data - theoretical insights

```python
# let's generate all possible situations (uniformly) and bet on everything
bets = []
for p in np.arange(0.1,1,0.1):
    for b in np.arange(0.1,1,0.1):
        bets.append([p,1-p,b,1-b,p/b-1, (1-p)/(1-b)-1, p/b + (1-p)/(1-b) - 2])
out2 = pd.DataFrame(bets, columns=['P_m1','P_m2','P_b1','P_b2','ev1','ev2','ev1+2']).sort_values('ev1+2')
out2
```

```python
# let's try to bet with smart (true) fractions - that of course leads to universal profit
out2['kellyEv1'] = out2['P_m1'] * out2['P_m1'] / out2['P_b1'] - out2['P_m1']
out2['kellyEv2'] = out2['P_m2'] * out2['P_m2'] / out2['P_b2'] - out2['P_m2']
out2['kellyEv'] = out2['kellyEv1'] + out2['kellyEv2']
out2['kelly-Unit'] = out2['kellyEv'] - out2['ev1+2']
out2
```

```python
# but we don't know the true probabilities, we know only the bookmaker's odds - that of course leads to nothing
out2['b-kellyEv1'] = out2['P_b1'] * out2['P_m1'] / out2['P_b1'] - out2['P_b1']
out2['b-kellyEv2'] = out2['P_b2'] * out2['P_m2'] / out2['P_b2'] - out2['P_b2']
out2['b-kellyEv'] = out2['b-kellyEv1'] + out2['b-kellyEv2']
out2['b-kelly-Unit'] = out2['b-kellyEv'] - out2['ev1+2']
out2
```

```python
# selecting the lower odds will cure the negative EV cases, but breaks all the rest
out2["betLowerOdds"] = np.where(out2['P_b1']>out2['P_b2'], out2['P_m1'], out2['P_m2']) / np.where(out2['P_b1']>out2['P_b2'], out2['P_b1'], out2['P_b2']) - 1
out2
```

```python
out2.sum()
```

### analysis
- EV is not linear w.r.t. the probabilities (of course, it's a division)! thus it does matter which way we move our mistakes!
- the total EV is negative in situations, when "bookmaker (slightly) underestimates the favorite"
    - because we lose more money on the overestimated outsider than we gain by the value-bet on favorite with the low odds
    - that explains a very good motivation for the bookmaker to set the odds a bit lower for the outsider (= long-shot bias) than the favorite
        - which has nothing to do with the bullshit psychological explanation (which can as well be used to the explain the contrary)
            - similarly to the log-utility function and Kelly, it's a nice math explanation of a psycho-phenomenon
- there are only 15% of the negative EV cases in a 2-outcome uniform distribution (unfortunatelly that appears to be the real situation case)
    - and it is impossible to distinguish the good bet (+EV) from a bad (-EV) bet purely from the constellation of the odds
    - or to find a good combination of fractions purely from the odds (is it really? maybe try a proof with a linear equation system)
        - we need a good model for that anyway! otherwise they are indistinguishible
            - even in this simple case without margin

```python
# how about 3 outcomes...
bets = []
for p1 in np.arange(0.1,1,0.1):
    for p2 in np.arange(0.1,1,0.1):
        if p1+p2 >=1:
            continue
        else:
            for b1 in np.arange(0.1,1,0.1):
                for b2 in np.arange(0.1,1,0.1):
                    if b1+b2 >=1:
                        continue
                    else:
                        bets.append([p1,p2,1-(p1+p2),b1,b2,1-(b1+b2),p1/b1-1, p2/b2-1, (1-(p1+p2))/(1-(b1+b2))-1, p1/b1 + p2/b2 + (1-(p1+p2))/(1-(b1+b2))-3])
out3 = pd.DataFrame(bets, columns=['P_m1','P_m2','P_m3','P_b1','P_b2','P_b3','ev1','ev2','ev3','ev1+2+3']).sort_values('ev1+2+3')

out3['diff'] = np.sum(out3[['P_m1','P_m2','P_m3']].values - out3[['P_b1','P_b2','P_b3']].values, axis=1)
out3
```

- there are only 11% of the negative EV cases but it's a similar story otherwise...

# Real data

- so following from the theoretical analysis, it is not automatically granted to beat a zero-margin bookmaker, but there should be a really good chance
    - an 89% chance to beat him, supposing that his predictions are distributed uniformly
        - which they are obviously not, but still a good prior...let's see

```python
bet_cols = ['BetH', 'BetD', 'BetA']
odds_cols = ['OddsH', 'OddsD', 'OddsA']
res_cols = ['W', 'D', 'L']
pred_cols = ['pH', 'pD', 'pA']
ev_cols = ['evH','evD','evA']
```

```python
# Ondra is negative with his strategy...ok, let's see some others
hist = pd.read_csv('./hist.csv', index_col=0).dropna()
hist.Date = pd.to_datetime(hist.Date, cache=True)
hist.rename(columns={'Profit' : "betOndraProfit"}, inplace=True)
hist['evOndra'] = (hist[pred_cols].values * hist[odds_cols].values * hist[bet_cols].values).sum(axis=1) - 1
hist.betOndraProfit.sum(), hist.betOndraProfit.mean()
```

## Distorting the odds/margin
- this works as expected, equalizing EVs of minOdds and MaxOdds
  - it looks quite good, except the betH strategy, which is still very positive
- but the bookmaker seems biased statically towards Home (instead of favorite)

```python
n_outcomes = 3
margin = 0.001
# add back some margin
hist[odds_cols] = hist[odds_cols] * (1 - margin)

inverted_odds = 1 / hist[odds_cols]
inverted_odds_sum = np.sum(inverted_odds, axis=1)
inverted_odds_sum = inverted_odds_sum[:, None]
margins = (inverted_odds_sum - 1) / (inverted_odds_sum)    # corrected margin definition

# HYPOTHESIS - margin is proportional to the odds (from Buchdahl) -> decrease the EV of the favorite
hist[odds_cols] = (n_outcomes * hist[odds_cols]) / (n_outcomes - (margins * hist[odds_cols]))
hist[["marginH","marginD","marginA"]] = (margins * hist[odds_cols]) / n_outcomes
hist
```

```python
hist['oddsNormBefore'] = (1/hist[odds_cols]).sum(axis=1)
hist[odds_cols] = hist[odds_cols].mul(hist['oddsNormBefore'],axis=0)
hist['oddsNormAfter'] = (1/hist[odds_cols]).sum(axis=1)
data = hist[['W','D','L','OddsH','OddsD','OddsA','oddsNormBefore','oddsNormAfter','pH','pD','pA','BetH','BetD','BetA','evOndra','betOndraProfit']]
data
```

```python
evs = pd.DataFrame((hist[pred_cols].values * hist[odds_cols].values) - 1, columns = ev_cols, index= data.index)
data = data.join(evs)
data['evSum'] = data[ev_cols].sum(axis=1)
data.sort_values('evSum')
```

```python
# obviously, there is always at least 1 EV+ opportunity and at least 1 EV- opportunity, that's correct...
pluses = np.where(data[ev_cols] > 0, 1, 0).sum(axis=1)
np.where(pluses==0)
np.where(pluses==1)
np.where(pluses==2)
```

```python
# ..now let's try to find them with some ad-hoc betting strategies
data['betHProfit'] = (data['OddsH'].values * data['W'].values) - 1
data['betDProfit'] = (data['OddsD'].values * data['D'].values) - 1
data['betAProfit'] = (data['OddsA'].values * data['L'].values) - 1

data['betAllProfit'] = (data[odds_cols].values * data[res_cols].values).sum(axis=1) - 3

data['betAllKelly'] = (data[odds_cols].values * data[pred_cols].values * data[res_cols].values).sum(axis=1) - 1

data['betBestKelly'] = (data[odds_cols].values * data[pred_cols].values * data[res_cols].values * data[bet_cols].values).sum(axis=1) - (data[pred_cols].values * data[bet_cols].values).sum(axis=1)

data
```

```python
# how about random betting
from sklearn.preprocessing import label_binarize
rand_bet = label_binarize(np.random.choice(3,len(data)), classes=[0,1,2])
data['betRandProfit'] = (data[odds_cols].values * rand_bet * data[res_cols].values).sum(axis=1) - 1
```

```python
# wut? the profits don't look good...
data.sum()
```

```python
# random was positive, but that can be just a chance, AVG profit with random betting is very negative!
tries = 100
sum_ = 0
for i in range(tries):
    rand_bet = label_binarize(np.random.choice(3,len(data)), classes=[0,1,2])
    sum_ += ((data[odds_cols].values * rand_bet * data[res_cols].values).sum(axis=1) - 1).sum()
sum_/tries
```

```python
# and true kelly goes bankrupt!
bank = 1
for ind,bet in data['betAllKelly'].iteritems():
    bank += bet*bank
#     print(bank)
bank
```

```python
# let's see profit in the directions of EV
sorted_ = np.argsort(data[ev_cols].values)
# (stupid python, so you're supposed to know a name of every function/field from every library? lol..)
data['betMaxEvProfit'] = (data[odds_cols].values * data[res_cols].values)[range(len(data)),sorted_[:,2]] - 1
data['betMidEvProfit'] = (data[odds_cols].values * data[res_cols].values)[range(len(data)),sorted_[:,1]] - 1
data['betMinEvProfit'] = (data[odds_cols].values * data[res_cols].values)[range(len(data)),sorted_[:,0]] - 1

posEv = np.where(data[ev_cols].values>0,1,0)
data['betPosEvProfit'] = (data[odds_cols].values * posEv * data[res_cols].values).sum(axis=1) - np.sum(posEv, axis=1)
```

```python
# let's see profit in directions of Odds
sorted_odds = np.argsort(data[odds_cols].values)
data['betMaxOddsProfit'] = (data[odds_cols].values * data[res_cols].values)[range(len(data)),sorted_odds[:,2]] - 1
data['betMidOddsProfit'] = (data[odds_cols].values * data[res_cols].values)[range(len(data)),sorted_odds[:,1]] - 1
data['betMinOddsProfit'] = (data[odds_cols].values * data[res_cols].values)[range(len(data)),sorted_odds[:,0]] - 1
```

```python
# again, not so good, but there are some interesting insights...
data.sum()
```

## backward analysis
- the real odds are apparently exactly within the unlucky 11% case! That bastard bookmaker...
    - assuming he has the same amount of money laid on all the opportunities, this gives him the best EV
        - and the worst negative EV to us in case we bet everything (to the bettor in aggregation)
- but it also makes him vulnerable if we just bet on the favorite/home (actually Home is even better than favorite, but that could be some artifact)
    - which is impossible for everyone to do in reality, because he would move the odds then...
